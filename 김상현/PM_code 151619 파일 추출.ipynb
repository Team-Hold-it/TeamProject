```python
import os
import json

folder_path = r'CCTV 라벨링 모아둔 폴더 경로'

# 파일 이름을 저장할 리스트
matching_files = []

# 경로 안의 모든 파일 검색
for root, dirs, files in os.walk(folder_path):
    for file in files:
        if file.endswith('.json'):
            file_path = os.path.join(root, file)
            with open(file_path, 'r') as json_file:
                data = json.load(json_file)
                pm_annotations = data.get('annotations', {}).get('PM', [])
                for annotation in pm_annotations:
                    pm_code = annotation.get('PM_code', '')
                    if '19' in pm_code: #조건문으로 다른 PM_code 추가가능
                        file_name = os.path.splitext(file)[0]
                        matching_files.append(file_name)

# 결과 출력
print(matching_files)
```

matching_files(특정 PM_code) 리스트에 있는 파일명을 포함한 라벨링 데이터 복제 후 이동

```python
import os
import json
import shutil

source_folder = r'path'
target_folder = r'path'


# 각 폴더에 대해 작업 수행
for folder_name in ['train']:
    source_path = os.path.join(source_folder, folder_name)
    target_path = os.path.join(target_folder, folder_name)

    # 대상 폴더가 없다면 생성
    if not os.path.exists(target_path):
        os.makedirs(target_path)

    # 소스 폴더 내의 파일 검색
    for root, dirs, files in os.walk(source_path):
        for file in files:
            if file.endswith('.txt'):
                file_name = os.path.splitext(file)[0]

                # matching_files에 들어간 파일 이름을 포함하는지 확인
                if any(match_file in file_name for match_file in matching_files):
                    source_file_path = os.path.join(root, file)
                    target_file_path = os.path.join(target_path, file)
                    
                    # 파일 복제
                    shutil.copy(source_file_path, target_file_path)

# 작업 완료
print("파일 복제 완료")
```
